```python
def random_sleep(t, t2):
    sleep_time = t+(t2*random.random())
    print('\n', sleep_time, '秒スリープ -> ', end='')
    time.sleep(sleep_time) # サーバーに負担をかけないために間隔をあける
    print('復帰\n')
```
