# ヤフーからサムネイル画像を取得するプログラムです

```python
import requests
from bs4 import BeautifulSoup
import time
import random
import lxml
import uuid
```

## アクセス

```python
session = requests.session()

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36',
    'Accept-Language': 'ja,en-US;q=0.9,en;q=0.8'
}

# URL
#url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-01&auccat=2084258814&va=ddh-01&b=1&n=20&rewrite_category=1"
url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-01&auccat=2084258814&va=ddh-01&b=1&n=100&rewrite_category=1"
#url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-01&auccat=2084258814&va=ddh-01&b=101&n=100&rewrite_category=0"
#url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-01&auccat=2084258814&va=ddh-01&b=201&n=100&rewrite_category=0"

req = session.get(url, headers = headers) # 取得
soup = BeautifulSoup(req.text, 'lxml') # 変換
products = soup.find_all('li', class_='Product')
```

# スクレイピング

```python
for i in [1, 101, 201]:
    print(i, '～')
    #url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-01&auccat=2084258814&va=ddh-01&b="+str(i)+"&n=100&rewrite_category=1"
    #url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-10&va=ddh-10&b="+str(i)+"&n=100&rewrite_category=1"
    url = "https://auctions.yahoo.co.jp/closedsearch/closedsearch?p=ddh-09&auccat=2084258814&va=ddh-09&b="+str(i)+"&n=100&rewrite_category=0"
    req = session.get(url, headers = headers) # データを取得
    soup = BeautifulSoup(req.text, 'lxml') # 変換
    products = soup.find_all('li', class_='Product') # 商品情報を取得

    for product, j in zip(products[1:], range(len(products))):
        print(i + j, "番目")
        price = product.find('span', class_='Product__priceValue').text.replace('円', '').replace(',', '')# 価格
        print(price)
        img_url = product.find('img', class_='Product__imageData')['src'] # 画像のURL
        print(img_url)
        r = session.get(img_url)
        with open(str('./scraped_img/original_data_ddh-09/')+ price.zfill(8) + '_' +str(uuid.uuid4())+str('.jpeg'), 'wb') as file:
            file.write(r.content)
        time.sleep(0.1+(1+random.random())) # サーバーに負担をかけないために間隔をあける
    time.sleep(3+(2+random.random())) # サーバーに負担をかけないために間隔をあける
```
